## 🤖 Test AI Functionality

```python
# Test text processing with transformers
from transformers import pipeline

# Create a sentiment analysis pipeline
print("Loading sentiment analysis model...")
classifier = pipeline("sentiment-analysis")

# Test it
texts = [
    "I love developing AI applications!",
    "Docker makes development so much easier",
    "Apple Silicon performance is amazing"
]

print("\nAnalyzing sentiment:")
for text in texts:
    result = classifier(text)[0]
    print(f"'{text}' -> {result['label']} ({result['score']:.3f})")
```

```python
# Test embeddings with sentence transformers
from sentence_transformers import SentenceTransformer
import numpy as np

print("Loading sentence transformer model...")
model = SentenceTransformer('all-MiniLM-L6-v2')

# Create embeddings
sentences = [
    "AI development with Docker",
    "Machine learning on Apple Silicon",
    "Building neural networks",
    "Data science workflows"
]

embeddings = model.encode(sentences)
print(f"\nCreated embeddings: {embeddings.shape}")

# Calculate similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(embeddings)

print("\nSimilarity matrix:")
for i, sent1 in enumerate(sentences):
    for j, sent2 in enumerate(sentences):
        if i < j:  # Only show upper triangle
            sim = similarity_matrix[i][j]
            print(f"'{sent1}' <-> '{sent2}': {sim:.3f}")
```

## 📊 Data Visualization

```python
# Create a simple visualization
import matplotlib.pyplot as plt
import numpy as np

# Generate sample AI model performance data
models = ['GPT-2', 'BERT', 'RoBERTa', 'T5', 'DistilBERT']
accuracy = [0.85, 0.91, 0.93, 0.89, 0.87]
speed = [120, 45, 40, 80, 95]  # inference time in ms

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Accuracy chart
ax1.bar(models, accuracy, color='lightblue')
ax1.set_title('Model Accuracy Comparison')
ax1.set_ylabel('Accuracy')
ax1.set_ylim(0, 1)
ax1.tick_params(axis='x', rotation=45)

# Speed chart
ax2.bar(models, speed, color='lightcoral')
ax2.set_title('Model Inference Speed')
ax2.set_ylabel('Time (ms)')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print("✅ Visualization test complete!")
```
